## Importing Pandas Library

```python
import pandas as pd
```

## Importing data file in the Python environment
Change the Windows folder location as per your system 

```python
#### Importing Expense File
imp_data = pd.read_csv("C:\\Ujjwal\\Analytics\\Git\\Scenario_2\\expenses.csv")
imp_data.head(2)
```

## Creating Long Form of the Expense Table

```python
#### Converting data in the long form
Long_data = pd.melt(imp_data, id_vars=["Year", "Month"], 
                    value_vars=["Grocery", "Travel", "Entertainment"], 
                    var_name="Expense Category", 
                    value_name="Expense Value")
Long_data.head()
```

## Monthly expenses in different expense categories 

```python
pd.pivot_table(Long_data, index=["Year", "Expense Category"], columns="Month", values="Expense Value").reset_index()
```
